```python

from google.colab import files
uploaded = files.upload()

# Step 4: Load the Excel file (replace with your uploaded filename)
file_name = list(uploaded.keys())[0]
xl = pd.ExcelFile(file_name)

# Optional: See all sheet names
print("Available Sheets:", xl.sheet_names)

```

```python

df = xl.parse(xl.sheet_names[0])

# Step 6: Display first few rows
df.head()


```

```python

columns_to_keep = [
    'Order ID', 'Product ID', 'Category', 'Sub-Category',
    'Sales', 'Profit', 'Quantity', 'Region', 'Order Date'
]
df_clean = df[columns_to_keep].copy()

```

```python


# Step 8: Convert date column to datetime
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'], errors='coerce')

# Step 9: Remove rows with missing values
df_clean.dropna(inplace=True)

# Step 10: Remove invalid sales/profit (optional)
df_clean = df_clean[(df_clean['Sales'] >= 0) & (df_clean['Profit'].notnull())]

# Step 11: Preview cleaned data
print("Cleaned Data Preview:")
print(df_clean.head())

# Step 12: Export to CSV
df_clean.to_csv("cleaned_superstore.csv", index=False)
files.download("cleaned_superstore.csv")
```
